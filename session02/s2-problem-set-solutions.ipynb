```python
#a)
n = 10**5 
p = (10**4) / (36**6)
rv = binom(n, p)
```

```python
#b)
noHits = binom.pmf(0, n, p)
print("Probability of no hits: ", noHits)
```

```python
#c)
mean = binom.mean(n, p)
print("Mean: ", mean)

varience = binom.var(n, p)
print("Varience: ", varience)
```

```python
#a)
n = 125
p = 0.9
showsUpHaveASeat = binom.cdf(120, n, p)
print("Probability of 120 or more people showing up and having a seat: ", showsUpHaveASeat)
```

```python
#b)
atLeastOneFreeSeat = binom.cdf(119, n, p)
print("Probability of at least one free seat: ", atLeastOneFreeSeat)
```
